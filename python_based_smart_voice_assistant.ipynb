```python
# Install necessary packages
!pip install SpeechRecognition
!pip install pyttsx3
!pip install pywhatkit
!apt-get update
!apt-get install -y xvfb
!pip install pyvirtualdisplay



import speech_recognition as sr
import pyttsx3
import datetime
import wikipedia
from pyvirtualdisplay import Display
import os

display = None

try:

    print("Starting virtual display...")
    display = Display(visible=0, size=(800, 600))

    display.start()
    print(f"Virtual display started: {display}")


    os.environ['DISPLAY'] = f":{display.display}"
    print(f"Setting DISPLAY environment variable to: {os.environ['DISPLAY']}")



    import pywhatkit


    r = sr.Recognizer()


    try:

        machine = pyttsx3.init('espeak')
        print("pyttsx3 initialized with espeak.")
    except Exception as e:
        print(f"Could not initialize pyttsx3 with espeak: {e}")
        print("Attempting initialization without specifying driver.")
        try:
            machine = pyttsx3.init()
            print("pyttsx3 initialized with default driver.")
        except Exception as e_fallback:
            print(f"Could not initialize pyttsx3: {e_fallback}")
            machine = None
            print("pyttsx3 initialization failed.")



    def talk(text):
         if machine:
             machine.say(text)
             machine.runAndWait()
         else:
             print(f"Text-to-speech engine not initialized. Cannot say: '{text}'")

    def get_instruction():
        instruction = ""
        try:

           with sr.Microphone() as source:
              print("Listening...")
              r.adjust_for_ambient_noise(source, duration=5)
              print("Say something!")

              speech = r.listen(source, timeout=10, phrase_time_limit=5)


              print("Recognizing...")
              instruction = r.recognize_google(speech)
              instruction = instruction.lower()
              print(f"Heard: {instruction}")


              if "robo" in instruction:

                 return instruction.replace('robo', "").strip()

              return instruction.strip()


        except sr.UnknownValueError:
            print("Google Speech Recognition could not understand audio")
            instruction = ""
        except sr.WaitTimeoutError:
            print("Listening timed out while waiting for phrase to start or duration")
            instruction = ""
        except sr.RequestError as e:
            print(f"Could not request results from Google Speech Recognition service; {e}")
            instruction = ""
        except Exception as e:

            print(f"An unexpected error occurred during speech recognition: {e}")
            instruction = ""


        return instruction.strip()



    def play_Robo():

        instruction = get_instruction()
        print(f"Processing instruction: '{instruction}'")

        if instruction and "robo" in instruction:

            processed_instruction = instruction.replace('robo', '').strip()
            print(f"Cleaned instruction after 'robo' removal: '{processed_instruction}'")

            if not processed_instruction:
                talk("Yes? How can I help?")
                print("Heard 'robo', but no command followed.")
                return

            if "play" in processed_instruction:

               song = processed_instruction.replace('play', "").strip()
               if song:
                   talk("Playing " + song)

                   try:
                       pywhatkit.playonyt(song)

                   except Exception as e:
                       print(f"Could not play on YouTube: {e}")
                       talk("Sorry, I could not play that on YouTube.")
               else:
                   talk("Please tell me what song to play.")
                   print("No song specified to play.")



            elif 'time' in processed_instruction:
                time = datetime.datetime.now().strftime('%I:%M %p')
                talk('The current time is ' + time)
                print("The current time is " + time)

            elif 'date' in processed_instruction:
                date = datetime.datetime.now().strftime('%d /%m /%Y')
                talk("Today's Date" + date)
                print("Today's date is " + date)


            elif 'how are you' in processed_instruction:
                 talk("I am fine, How about you?")
                 print("I am fine, How about you?")

            elif 'what is your name' in processed_instruction:
                talk("I am Robo, What can I do for you?")
                print("I am Robo, What can I do for you?")

            elif 'who is' in processed_instruction:

                command = processed_instruction.replace("who is", "").strip()
                if command:
                    try:

                        info = wikipedia.summary(command, sentences=1)
                        print(info)
                        talk(info)
                    except wikipedia.exceptions.PageError:
                         talk("Sorry, I couldn't find information on that.")
                         print(f"Wikipedia page not found for '{command}'.")
                    except wikipedia.exceptions.DisambiguationError as e:

                         options_preview = ", ".join(e.options[:5]) + ("..." if len(e.options) > 5 else "")
                         talk(f"Could you be more specific? Options include: {options_preview}")
                         print(f"Wikipedia disambiguation error for '{command}'. Options: {e.options}")
                    except Exception as e:
                         print(f"An error occurred while fetching Wikipedia info for '{command}': {e}")
                         talk("Sorry, I encountered an error while fetching that information.")
                else:
                     talk("Please tell me who you want information about.")
                     print("No subject specified for Wikipedia search.")


            elif 'stop' in processed_instruction or 'exit' in processed_instruction:
                 talk("Goodbye!")
                 print("Stopping Robo.")
                 return



            else:
                talk("Sorry, I didn't understand that command. Please repeat or try something else.")
                print(f"Unrecognized command: '{processed_instruction}'")

        elif instruction and "robo" not in instruction:

             talk("Please say 'Robo' before your command.")
             print(f"Instruction received but did not contain 'robo': '{instruction}'")

        else:

             talk("I didn't hear anything or could not understand. Please try again.")
             print("No valid instruction received or recognized.")



    play_Robo()

except Exception as e:

    print(f"An unexpected error occurred outside the main loop: {e}")

finally:

    if display:
         print("Stopping virtual display.")
         display.stop()
```
